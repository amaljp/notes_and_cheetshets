```python
#selecting the homogeneous data

colleges = pd.read_csv('data/college.csv', index_col='INSTNM') #label index col as instnm: institutin name
```

```python
college_ugds = colleges.filter(like="UGDS_") #filterning wih 
college_ugds.head()
```

```python
college_ugds.head(2)
```

```python
(
    college_ugds.isnull()
    .sum(axis=1) #for each row
    .sort_values(ascending=False) #descending
    .head(2)

)
```

```python
college_ugds = college_ugds.dropna(how='all')
college_ugds.isna().sum()
```

```python
diversity_metric = college_ugds.ge(0.15).sum(axis=1) #more than 15% filter
diversity_metric.head()
```

```python
diversity_metric.value_counts() #5 means highly diverse? ie, 
```

```python
diversity_metric.sort_values(ascending=False).head()
```
