```python
movie = pd.read_csv('data/movie.csv')
```

```python
movie2 = movie[["movie_title", "imdb_score", "budget"]]
```

```python
top100 = movie2.nlargest(100, "imdb_score")
```

```python
top100.nsmallest(5, 'budget')
```

```python
import matplotlib.pyplot as plt
```

```python

fig = plt.figure()
```

```python
x = np.linspace(0, 10, 100)

plt.figure()
plt.plot(x, np.sin(x), '-')
plt.plot(x, np.cos(x), '--');
```

```python
#matlab interface

plt.subplot(2,1,1) #rows, columns, pannel no.
plt.plot(x, np.sin(x))

plt.subplot(2,1,2)
plt.plot(x, np.cos(x));
```

```python
#obj oriented interface.

fig, ax = plt.subplots(2)

ax[0].plot(x, np.sin(x))
ax[1].plot(x, np.cos(x));
```
